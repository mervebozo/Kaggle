## Exploratory Data Analysis (EDA)

This notebook aims to analyze and visualize the Roosevelt National Forest dataset.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
plt.style.use('ggplot')
rcParams['figure.figsize'] = 16, 10
cmap = sns.diverging_palette(220, 10, as_cmap=True)
```

```python
# load the data
df_train = pd.read_csv('./input/train.csv', index_col='Id')
df_test  = pd.read_csv('./input/test.csv', index_col='Id')
```

```python
print("Train shape: ", df_train.shape)
print("Test shape: ", df_test.shape)
```

```python
df_train.iloc[:, 10:-1] = df_train.iloc[:, 10:-1].astype("category")
df_test.iloc[:, 10:] = df_test.iloc[:, 10:].astype("category")
```

```python
# box and whisker plots
rcParams['figure.figsize'] = 16, 14
df_train.plot(kind='box', subplots=True, layout=(4, 4), sharex=False, sharey=False)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="white")
f, ax = plt.subplots(figsize=(15, 15))

corr = df_train.corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, cmap=cmap, vmax=.4, center=0,
            square=True, linewidths=.6, mask=mask, annot=True, fmt=".3f", cbar_kws={"shrink": .5})
plt.title('Pearson Correlation of Features', y=1.05, size=15)
plt.show()
```

It can be seen there are some positively related features like Horizontal_Distance_To_Hydrology-Vertical_Distance_To_Hydrology, 
Hillshade_3pm-Aspect, Hillshade_3pm-Hillshade_Noon and some negatively correlate features like Hillshade_Noon-Slope, Hillshade_9am-Hillshade_3pm.

We can either check the feature importance graph and make a decision of one of the most correlated feature drop and another to keep.

```python
rcParams['figure.figsize'] = 12, 14
from sklearn.ensemble import RandomForestClassifier

df_train_copy = df_train.copy()
y_train = df_train_copy.pop('Cover_Type')
x_train = df_train_copy.copy()

model=RandomForestClassifier()
model.fit(x_train, y_train)
pd.Series(model.feature_importances_,x_train.columns).sort_values(ascending=True).plot.barh(width=0.7,  color='navy')
plt.title('Feature Importance in Random Forests')
plt.tight_layout()
plt.show()
```

```python
df_train.columns
```

We can see the feature importance for the classifier RandomForests.
* Some of the common important features are Elevation, Horizontal_Distance_To_Roadways, Horizontal_Distance_To_Fire_Points, Horizontal_Distance_To_Hydrology, Aspect, Vertical_Distance_To_Hydrology, Hillshade_9am and Wilderness_Area4.

* The features after Slope can be dropped based on this feature importance graph.
